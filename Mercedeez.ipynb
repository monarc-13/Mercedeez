```python
import pandas as pd
df_train = pd.read_excel('Train_Mercedeez.xlsx')
df_train
```

```python
df_train.isnull().sum()
```

```python
y_train = df_train['y'].values
```

```python
#This line of code is written in Python and it is using list comprehension to create a new list called "cols". It iterates through the columns in the dataframe "df_train" and selects only those columns where the column name contains the letter 'X'.
#So, the resulting "cols" list will contain only the columns from the dataframe "df_train" that have the letter 'X' in their names. This can be useful for selecting specific columns based on a certain condition or pattern.
cols = [c for c in df_train.columns if 'X' in c]
print('Number of features: {}'.format(len(cols))) #This will display the number of columns in the dataframe "df_train" that contain the letter 'X'

print('Feature types:')
df_train[cols].dtypes.value_counts()
```

```python
counts = [[], [], []]
for c in cols:
    typ = df_train[c].dtype
    uniq = len(np.unique(df_train[c]))
    if uniq == 1:
        counts[0].append(c)
    elif uniq == 2 and typ == np.int64:
        counts[1].append(c)
    else:
        counts[2].append(c)
```

```python
print('Constant features: {} Binary features: {} Categorical features: {}\n'
      .format(*[len(c) for c in counts]))
print('Constant features:', counts[0])
print('Categorical features:', counts[2])
```

```python
df_test = pd.read_excel('test_Merce.xlsx')
df_test
```

```python
#It creates a list of columns in the dataset (df_train) that are not the 'ID' and 'y' columns. 'set()' is used to remove duplicate values from the columns, and then 'list()' is used to convert
#them into a list is called usable\_columns. The usable\_columns list will contain all the unique columns in the dataset that are not used for target variable prediction.
usable_columns = list(set(df_train.columns) - set(['ID', 'y']))
y_train = df_train['y'].values
id_test = df_test['ID'].values

x_train = df_train[usable_columns]
x_test = df_test[usable_columns]
```

```python

for column in usable_columns:
    cardinality = len(np.unique(x_train[column]))
    if cardinality == 1:
        x_train.drop(column, axis=1) # Column with only one
        # value is useless so we drop it
        x_test.drop(column, axis=1)
    if cardinality > 2: # Column is categorical
        mapper = lambda x: sum([ord(digit) for digit in x])
        x_train[column] = x_train[column].apply(mapper)
        x_test[column] = x_test[column].apply(mapper)
x_train.head()
```

```python
# Step9: Make sure the data is now changed into numericals

print('Feature types:')
x_train[cols].dtypes.value_counts()
```

```python
# Step10: Perform dimensionality reduction
# Linear dimensionality reduction using Singular Value Decomposition of
# the data to project it to a lower dimensional space.
n_comp = 12
pca = PCA(n_components=n_comp, random_state=420)
pca2_results_train = pca.fit_transform(x_train)
pca2_results_test = pca.transform(x_test)
```

```python
# Step11: This code essentially sets up and trains an XGBoost regression model,
# monitoring its performance on both the training and validation sets using the R-squared metric. The early stopping mechanism helps
# prevent overfitting, and the training progress is printed to the console every 10 rounds.

import xgboost as xgb
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(
        pca2_results_train,
        y_train, test_size=0.2,
        random_state=4242)

d_train = xgb.DMatrix(x_train, label=y_train)
d_valid = xgb.DMatrix(x_valid, label=y_valid)
#d_test = xgb.DMatrix(x_test)
d_test = xgb.DMatrix(pca2_results_test)

params = {}
params['objective'] = 'reg:linear'
params['eta'] = 0.02
params['max_depth'] = 4

def xgb_r2_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)

watchlist = [(d_train, 'train'), (d_valid, 'valid')]

clf = xgb.train(params, d_train,
                1000, watchlist, early_stopping_rounds=50,
                feval=xgb_r2_score, maximize=True, verbose_eval=10)
```

```python

p_test = clf.predict(d_test)

sub = pd.DataFrame()
sub['ID'] = id_test
sub['y'] = p_test
sub.to_csv('xgb.csv', index=False)

sub.head()
```
